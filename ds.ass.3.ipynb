# 5. Scenario: A study is conducted to compare the performance of two different teaching methods. The test scores of the students in each group are as follows:
   Group A: [85, 90, 92, 88, 91]
   Group B: [82, 88, 90, 86, 87]
   Perform a t-test to determine if there is a significant difference in the mean scores between the two groups.


```python
Step 1: Define the null hypothesis (H0) and the alternative hypothesis (Ha).

Null hypothesis (H0): There is no significant difference in the mean scores between Group A and Group B.
Alternative hypothesis (Ha): There is a significant difference in the mean scores between Group A and Group B.
Step 2: Calculate the means of Group A and Group B.

Mean of Group A: (85 + 90 + 92 + 88 + 91) / 5 = 89.2
Mean of Group B: (82 + 88 + 90 + 86 + 87) / 5 = 86.6
Step 3: Calculate the variances of Group A and Group B.

Variance of Group A:
Calculate the squared differences from the mean for each score: (85-89.2)^2, (90-89.2)^2, (92-89.2)^2, (88-89.2)^2, (91-89.2)^2
Sum the squared differences: (85-89.2)^2 + (90-89.2)^2 + (92-89.2)^2 + (88-89.2)^2 + (91-89.2)^2 = 18.16 + 0.64 + 7.84 + 1.44 + 1.44 = 29.52
Divide the sum by (n-1), where n is the number of scores in Group A: 29.52 / (5-1) = 7.38
Variance of Group B:
Calculate the squared differences from the mean for each score: (82-86.6)^2, (88-86.6)^2, (90-86.6)^2, (86-86.6)^2, (87-86.6)^2
Sum the squared differences: (82-86.6)^2 + (88-86.6)^2 + (90-86.6)^2 + (86-86.6)^2 + (87-86.6)^2 = 18.76 + 1.96 + 7.84 + 0.16 + 0.04 = 28.76
Divide the sum by (n-1), where n is the number of scores in Group B: 28.76 / (5-1) = 7.19
Step 4: Calculate the pooled variance.

Pooled variance = [(nA - 1) * varianceA + (nB - 1) * varianceB] / (nA + nB - 2)
= [(5 - 1) * 7.38 + (5 - 1) * 7.19] / (5 + 5 - 2)
= (4 * 7.38 + 4 * 7.19) / 8
= 29.52 + 28.76 / 8
= 58.28 / 8
= 7.29
Step 5: Calculate the t-value.

t-value = (meanA - meanB) / sqrt(pooled variance * (1/nA + 1/nB))
= (89.2 - 86.6) / sqrt(7.29 * (1/5 + 1/5))
= 2.6 / sqrt(7.29 * 0.4)
= 2.6 / sqrt(2.916)
≈ 2.6 / 1.71
≈ 1.52
Step 6: Determine the critical t-value or p-value for the given significance level and degrees of freedom.

The critical t-value or p-value depends on the chosen significance level (e.g., α = 0.05) and the degrees of freedom (df = nA + nB - 2 = 5 + 5 - 2 = 8). You can consult a t-table or use statistical software to find the critical value.
Step 7: Compare the calculated t-value with the critical t-value or p-value.

If the calculated t-value is greater than the critical t-value or if the p-value is less than the chosen significance level (e.g., p < 0.05), then reject the null hypothesis.
If the calculated t-value is less than the critical t-value or if the p-value is greater than the chosen significance level, then fail to reject the null hypothesis.
```

# 9. Scenario: A study is conducted to compare the effectiveness of two different medications. The recovery times of the patients in each group are as follows:
#   Medication A: [10, 12, 14, 11, 13]
#   Medication B: [15, 17, 16, 14, 18]
#   Perform an analysis of variance (ANOVA) to determine if there is a significant difference in the mean recovery times between the two medications.



```python
Step 1: Calculate the mean of the recovery times for each group.

Mean of Medication A: (10 + 12 + 14 + 11 + 13) / 5 = 12
Mean of Medication B: (15 + 17 + 16 + 14 + 18) / 5 = 16
Step 2: Calculate the sum of squares between groups (SSG).

SSG = (nA * (meanA - meanOverall)^2) + (nB * (meanB - meanOverall)^2)
where nA and nB are the number of observations in Medication A and Medication B respectively, meanA and meanB are the means of Medication A and Medication B respectively, and meanOverall is the overall mean of the recovery times.

SSG = (5 * (12 - 14)^2) + (5 * (16 - 14)^2)

SSG = 40

Step 3: Calculate the sum of squares within groups (SSW).

SSW = (sum of squares for Medication A) + (sum of squares for Medication B)
where the sum of squares for each group is calculated as the sum of squared deviations from the group mean.

Sum of squares for Medication A = (10 - 12)^2 + (12 - 12)^2 + (14 - 12)^2 + (11 - 12)^2 + (13 - 12)^2

Sum of squares for Medication B = (15 - 16)^2 + (17 - 16)^2 + (16 - 16)^2 + (14 - 16)^2 + (18 - 16)^2

Sum of squares for Medication A = 10 + 0 + 2 + 1 + 1 = 14

Sum of squares for Medication B = 1 + 1 + 0 + 4 + 4 = 10

SSW = 14 + 10 = 24

Step 4: Calculate the degrees of freedom between groups (dfG) and within groups (dfW).

dfG = number of groups - 1 = 2 - 1 = 1
dfW = total number of observations - number of groups = 10 - 2 = 8
Step 5: Calculate the mean squares between groups (MSG) and within groups (MSW).

MSG = SSG / dfG = 40 / 1 = 40
MSW = SSW / dfW = 24 / 8 = 3
Step 6: Calculate the F-statistic.

F-statistic = MSG / MSW = 40 / 3 = 13.33 (rounded to two decimal places)
Step 7: Determine the critical value of F for a given significance level (alpha) and degrees of freedom (dfG and dfW). Compare the calculated F-statistic to the critical value.

Step 8: If the calculated F-statistic is greater than the critical value, reject the null hypothesis and conclude that there is a significant difference in the mean recovery times between Medication A and Medication B. If the calculated F-statistic is not greater than the critical value, fail to reject the null hypothesis and conclude that there is not enough evidence to suggest a significant difference.
```

# 11. Scenario: A quality control department wants to test the weight consistency of a product. The weights of a sample of products are as follows:
#    [10.2, 9.8, 10.0, 10.5, 10.3, 10.1]
#    Perform a hypothesis test to determine if the mean weight differs significantly from 10 grams.


```python
To perform a hypothesis test to determine if the mean weight differs significantly from 10 grams, you can use a t-test. Here are the steps to conduct the test:

Step 1: State the null hypothesis (H0) and the alternative hypothesis (H1):

H0: The mean weight is equal to 10 grams.
H1: The mean weight is not equal to 10 grams.

Step 2: Choose the significance level (α) for the test. Commonly used values are 0.05 or 0.01, representing a 5% or 1% level of significance, respectively.

Step 3: Calculate the sample mean (x̄) and the sample standard deviation (s) of the weights:

Sample mean (x̄) = (10.2 + 9.8 + 10.0 + 10.5 + 10.3 + 10.1) / 6 = 10.1667
Sample standard deviation (s) = √[((10.2 - x̄)² + (9.8 - x̄)² + (10.0 - x̄)² + (10.5 - x̄)² + (10.3 - x̄)² + (10.1 - x̄)²) / (6 - 1)]

Step 4: Calculate the t-value:

t-value = (x̄ - μ) / (s / √n)
where μ is the hypothesized population mean (10 grams) and n is the sample size.

t-value = (10.1667 - 10) / (s / √6)

Step 5: Determine the degrees of freedom (df), which is equal to the sample size minus 1:

df = 6 - 1 = 5

Step 6: Look up the critical t-value corresponding to the chosen significance level and the degrees of freedom. Compare the calculated t-value from Step 4 with the critical t-value to make a decision.

If the calculated t-value falls within the critical region (rejects the null hypothesis), then the mean weight differs significantly from 10 grams. Otherwise, if the calculated t-value does not fall within the critical region (fails to reject the null hypothesis), then there is not enough evidence to conclude a significant difference.
```

```python
To perform a chi-square test to determine if there is a significant difference in the click-through rates between Design A and Design B, you can follow these steps:

Step 1: State the null hypothesis (H0) and the alternative hypothesis (H1):

H0: There is no significant difference in the click-through rates between Design A and Design B.
H1: There is a significant difference in the click-through rates between Design A and Design B.

Step 2: Set the significance level (α) for the test. Commonly used values are 0.05 or 0.01, representing a 5% or 1% level of significance, respectively.

Step 3: Create a contingency table to summarize the observed frequencies:

css
Copy code
                  Design A   Design B
    Clicked           100         80
    Not Clicked       100         120
Step 4: Calculate the expected frequencies under the assumption of no significant difference. This can be done by calculating the row totals, column totals, and the total number of observations:

Row Total for Clicked: 100 + 80 = 180
Row Total for Not Clicked: 100 + 120 = 220
Column Total for Design A: 100 + 100 = 200
Column Total for Design B: 80 + 120 = 200
Total Number of Observations: 400

Expected Frequency for Design A and Clicked: (180 * 200) / 400 = 90
Expected Frequency for Design A and Not Clicked: (220 * 200) / 400 = 110
Expected Frequency for Design B and Clicked: (180 * 200) / 400 = 90
Expected Frequency for Design B and Not Clicked: (220 * 200) / 400 = 110

Step 5: Calculate the chi-square test statistic:

χ² = ∑ ((Observed Frequency - Expected Frequency)² / Expected Frequency)

χ² = ((100 - 90)² / 90) + ((80 - 90)² / 90) + ((100 - 110)² / 110) + ((120 - 110)² / 110)

Step 6: Determine the degrees of freedom (df), which is equal to (number of rows - 1) * (number of columns - 1):

df = (2 - 1) * (2 - 1) = 1

Step 7: Look up the critical chi-square value corresponding to the chosen significance level and the degrees of freedom. Compare the calculated chi-square value from Step 5 with the critical chi-square value to make a decision.

If the calculated chi-square value is greater than the critical chi-square value, then reject the null hypothesis and conclude that there is a significant difference in the click-through rates between Design A and Design B. Otherwise, if the calculated chi-square value is less than or equal to the critical chi-square value, fail to reject the null hypothesis and conclude that there is no significant difference.
```
